```python
df2['ADDRESS'] = df2["ADDRESS"].astype("category")# astype는 타입변경함수
df2['BEDROOMS'] = df2["BEDROOMS"].astype("category")
df2['BATHROOMS'] = df2["BATHROOMS"].astype("category")
df2['GARAGE'] = df2["GARAGE"].astype("category")
df2['LAND_AREA'] = df2["LAND_AREA"].astype("category")
df2['FLOOR_AREA'] = df2["FLOOR_AREA"].astype("category")
df2['BUILD_YEAR'] = df2["BUILD_YEAR"].astype("category")
df2['CBD_DIST'] = df2["CBD_DIST"].astype("category")
df2['NEAREST_STN '] = df2["NEAREST_STN"].astype("category")
df2['NEAREST_STN_DIST'] = df2["NEAREST_STN_DIST"].astype("category")
df2['DATE_SOLD'] = df2["DATE_SOLD"].astype("category")
df2['NEAREST_SCH'] = df2["NEAREST_SCH"].astype("category")
df2['NEAREST_SCH_DIST'] = df2["NEAREST_SCH_DIST"].astype("category")
df2['NEAREST_SCH_RANK '] = df2["NEAREST_SCH_RANK"].astype("category")
```

```python
from sklearn import model_selection
from sklearn. metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn import linear_model, datasets

# 학습데이터와 테스트데이터 분리
x_train, x_test, y_train, y_test = model_selection.train_test_split(x_data, y_data, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import Regression

classifiter = Regression
classifiter.fit(x_train, y_train)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x_train)
X_train_scaled = scaler.transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
model = linear_model.Regression(solver = "liblinear")
model.fit(X_train_scaled, y_train)
```

```python
pred = model.predict(X_test_scaled)
score = accuracy_score(y_test, pred)
print('test score: ', score)
```
