```python
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error , r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import scale
from sklearn.impute import SimpleImputer
df = pd.read_csv('./DATA/train.csv')
```

```python
imp_mean = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imp_median = SimpleImputer(missing_values = np.nan, strategy = 'median')
```

```python
df_garage = imp_mean.fit_transform(ndf[['GARAGE']])
df_garage
```

```python
df_BuildYear = imp_median.fit_transform(ndf[['BUILD_YEAR']])
df_BuildYear
```

```python
df_NearestSchRank = imp_mean.fit_transform(ndf[['NEAREST_SCH_RANK']])
df_NearestSchRank
```

```python
ndf['GARAGE'] = df_garage
ndf['BUILD_YEAR'] = df_BuildYear
ndf['NEAREST_SCH_RANK']= df_NearestSchRank
```

```python
ndf.head(10)
```

```python
#  
df_address = ndf['ADDRESS']
df_suburb = ndf['SUBURB']
df_nearest_stn = ndf['NEAREST_STN']
df_date_sold = ndf['DATE_SOLD']
df_NearestSch = ndf['NEAREST_SCH']
```

```python
le = LabelEncoder()
le.fit(df_address) 
address = le.transform(df_address)
```

```python
le.fit(df_suburb) 
suburb = le.transform(df_suburb)
```

```python
le.fit(df_nearest_stn) 
nearest_stn = le.transform(df_nearest_stn)
```

```python
le.fit(df_date_sold) 
date_sold = le.transform(df_date_sold)
```

```python
le.fit(df_NearestSch) 
NearestSch = le.transform(df_NearestSch)
```

```python
ndf['ADDRESS'] = address
ndf['SUBURB'] = suburb
ndf['NEAREST_STN'] = nearest_stn
ndf['DATE_SOLD'] = date_sold
ndf['NEAREST_SCH'] = NearestSch
```

```python
ndf
```

```python
# 
y_target = ndf['PRICE']
X_data = ndf.drop(['PRICE'],axis=1,inplace=False)
X_train , X_test , y_train , y_test = train_test_split(X_data , y_target , random_state=156)
```

```python
model_ndf = LinearRegression().fit(X_train, y_train)
```

```python
model_ndf.score(X_train, y_train)
```
