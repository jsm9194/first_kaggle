```python
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error , r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import scale
from sklearn.impute import SimpleImputer
df = pd.read_csv('train.csv')
df
```

```python
df.corr()
```

```python
# 가격과의 상관관계가 음수인 것을 제거함
ndf= df.drop(['BUILD_YEAR','CBD_DIST', 'NEAREST_STN_DIST', 'POSTCODE','NEAREST_SCH' ,'LONGITUDE', 'NEAREST_SCH_DIST' , 'NEAREST_SCH_RANK'], axis=1)
ndf.head()
```

```python
ndf.corr()
```

```python
ndf.info() #GARAGE에 nan값이 있는 것을 확인
```

```python
# 결측값 확인 작업
ndf.isnull().sum()
```

```python
ndf['GARAGE'].mean()
```

```python
# GARAGE의 평균값으로 결측값 처리
imp_mean = SimpleImputer(missing_values = np.nan, strategy = 'mean')
df_garage = imp_mean.fit_transform(ndf[['GARAGE']])
df_garage
```

```python
ndf['GARAGE'] = df_garage
```

```python
# 인코딩 작업
ndf.info()
```

```python
df_address = ndf['ADDRESS']
df_suburb = ndf['SUBURB']
df_nearest_stn = ndf['NEAREST_STN']
df_date_sold = ndf['DATE_SOLD']
```

```python
le = LabelEncoder()
```

```python
le.fit(df_address) 
address = le.transform(df_address)
```

```python
le.fit(df_suburb) 
suburb = le.transform(df_suburb)
```

```python
le.fit(df_nearest_stn) 
nearest_stn = le.transform(df_nearest_stn)
```

```python
le.fit(df_date_sold) 
date_sold = le.transform(df_date_sold)
```

```python
ndf['ADDRESS'] = address
ndf['SUBURB'] = suburb
ndf['NEAREST_STN'] = nearest_stn
ndf['DATE_SOLD'] = date_sold
```

```python
ndf.info()
```

```python
# 분류작업
y_target = ndf['PRICE']
X_data = ndf.drop(['PRICE'],axis=1,inplace=False)
X_train , X_test , y_train , y_test = train_test_split(X_data , y_target , random_state=156)
```

```python
model_ndf = LinearRegression().fit(X_train, y_train)
```

```python
model_ndf.coef_
```

```python
model_ndf.intercept_
```

```python
model_ndf.score(X_train, y_train)
```
